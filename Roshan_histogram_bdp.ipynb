```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv('/content/trip_cnts.csv')

# Calculate 'business-month'
df['business-month'] = df['business'] + '-' + df['month'].astype(str)

# Plot histogram with custom color
plt.figure(figsize=(10, 6))
bars = plt.bar(df['business-month'], df['trip_count'], color='skyblue')
plt.xlabel('Business-Month')
plt.ylabel('Trip Counts')
plt.title('Histogram of Trip Counts by Business-Month')
plt.xticks(rotation=45, ha='right')
# Set y-axis scale to logarithmic
plt.yscale('log')

# Add values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv('/content/trip_cnts.csv')

# Calculate 'business-month'
df['business-month'] = df['business'] + '-' + df['month'].astype(str)

# Separate DataFrame for Lyft and Uber businesses
lyft_df = df[df['business'] == "Lyft"]
uber_df = df[df['business'] == "Uber"]

# Plotting subgraphs for Lyft and Uber
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Lyft subgraph
axs[0].bar(lyft_df['business-month'], lyft_df['trip_count'], color='skyblue')
axs[0].set_title('Lyft')
axs[0].set_ylabel('Trip Counts')
axs[0].tick_params(axis='x', rotation=45, labelsize=8)

# Uber subgraph
axs[1].bar(uber_df['business-month'], uber_df['trip_count'], color='salmon')
axs[1].set_title('Uber')
axs[1].set_ylabel('Trip Counts')
axs[1].tick_params(axis='x', rotation=45, labelsize=8)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
df_earning = pd.read_csv('/content/profits_month.csv')

# Sample data (assuming columns are named 'business', 'month', and 'total_profit' in the CSV)
# If the column names are different, replace them accordingly
businesses = df_earning['business']
months = df_earning['month']
profits = df_earning['total_profit']

# Define colors for each business
colors = {'Lyft': 'skyblue', 'Uber': 'salmon'}

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar([f"{business}-{month}" for business, month in zip(businesses, months)], profits, color=[colors[business] for business in businesses])
plt.xlabel('Business-Month')
plt.ylabel('Total Profit')
plt.title('Total Profit for Each Business in Each Month')
plt.xticks(rotation=45, ha='right')
plt.ticklabel_format(style='plain', axis='y')

# Add values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom')

plt.tight_layout()
plt.show()
```

Task 2 part-ii (Business wise)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv('/content/profits_month.csv')

# Define colors for each business
colors = {'Lyft': 'skyblue', 'Uber': 'salmon'}

# Plotting for each business
for business, color in colors.items():
    # Filter data for the current business
    business_df = df[df['business'] == business]

    # Extract months and profits for the current business
    months = business_df['month']
    profits = business_df['total_profit']

    # Plotting
    plt.figure(figsize=(8, 6))
    plt.bar(months, profits, color=color)
    plt.xlabel('Month')
    plt.ylabel('Total Profit')
    plt.title(f'Total Profit for {business} in Each Month')
    plt.xticks(months)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv('/content/earnings_month.csv')

# colours are in hexadecimal
custom_colors = ['#FF5733', '#33FFB2', '#3366FF', '#FF33E6', '#FFFF33']

# Plotting
plt.figure(figsize=(10, 6))
bars = []
for i, row in df.iterrows():
    bar = plt.bar(f"{row['business']}-{row['month']}", row['total_earnings'], color=custom_colors[i % len(custom_colors)])
    bars.append(bar)

plt.xlabel('Business-Month')
plt.ylabel("Driver's Earnings")
plt.title("Driver's Earnings for Each Business in Each Month")
plt.xticks(rotation=45, ha='right')

# Add values on top of each bar
for bar in bars:
    for b in bar:
        yval = b.get_height()
        plt.text(b.get_x() + b.get_width()/2, yval, round(yval, 2), va='bottom')

plt.yscale('log')  # Setting y-axis scale to logarithmic

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV file
data = pd.read_csv('/content/earnings_month.csv')

# Separate data for Lyft and Uber
lyft_data = data[data['business'] == 'Lyft'][['month', 'total_earnings']]
uber_data = data[data['business'] == 'Uber'][['month', 'total_earnings']]

# Plotting for Lyft
plt.figure(figsize=(8, 4))
plt.bar(lyft_data['month'], lyft_data['total_earnings'], color='skyblue')
plt.xlabel('Month')
plt.ylabel("Driver's Earnings")
plt.title("Driver's Earnings for Lyft in Each Month")
plt.tight_layout()
plt.show()

# Plotting for Uber
plt.figure(figsize=(8, 6))
plt.bar(uber_data['month'], uber_data['total_earnings'], color='lightgreen')
plt.xlabel('Month')
plt.ylabel("Driver's Earnings")
plt.title("Driver's Earnings for Uber in Each Month")
plt.tight_layout()
plt.show()
```
